```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mental_health_df =pd.read_csv("Survey.csv")
mental_health_df.head()
mental_health_df = mental_health_df.loc[(mental_health_df['tech_company'] =='Yes')]
mental_health_df

mental_health_df_rename = mental_health_df.rename(columns={"family_history": "Family History",
                                            "work_interfere" : "Mental Illness Interfere",
                                           "treatment" : "Seeking Treatment",
                                            "obs_consequence" : "Heard/Observed Consequences",
                                            "wellness_program": "Employee Wellness Program",
                                            "benefits" : "Employer Benefits",
                                            "seek_help" : "Employer Provide Resources",
                                            "care_options" : "Knowledge of Options",
                                            "tech_company" : "Work in Tech",
                                            "state" : "State"
                                             })
mental_health_df_rename
```

```python
mental_health_count = mental_health_df["family_history"].value_counts()
mental_health_count.plot(kind="pie",autopct='%2.2f%%')
plt.ylabel(None)
plt.title("Family History of Mental Illness")
```

```python
mental_health_yes_work_interfere_df = mental_health_df[mental_health_df["family_history"]=="Yes"].copy()
mental_health_yes_work_interfere_count = mental_health_yes_work_interfere_df["work_interfere"].value_counts()
mental_health_yes_work_interfere_count.plot(kind="pie",autopct='%2.2f%%', colors = ["royalblue", "darkorange", "limegreen", "red" ])
plt.ylabel(None)
plt.title("Family History of Mental Health and Work Interference")
```

```python
mental_health_no_work_interfere_df = mental_health_df[mental_health_df["family_history"]=="No"].copy()
mental_health_no_work_interfere_count = mental_health_no_work_interfere_df["work_interfere"].value_counts()
mental_health_no_work_interfere_count.plot(kind="pie",autopct='%2.2f%%', colors = ["royalblue", "red", "limegreen","darkorange"])
plt.ylabel(None)
plt.title("No Family History of Mental Health and Work Interference")
```

```python
mental_health_df["anonymity"]= mental_health_df['anonymity'].replace({"Don't know": 'No or unsure', 'No': 'No or unsure'})
```

```python
mental_health_count = mental_health_df["anonymity"].value_counts()
mental_health_count.plot(kind="pie",autopct='%2.2f%%')
plt.ylabel(None)
plt.title("Protection for Seeking Resources")
```

```python
mental_health_df["seek_help"]= mental_health_df['seek_help'].replace({"Don't know": 'No or unsure', 'No': 'No or unsure'})
```

```python
mental_health_seek_help_df = mental_health_df[mental_health_df["anonymity"]=="Yes"].copy()
mental_health_seek_help_count = mental_health_seek_help_df["seek_help"].value_counts()
mental_health_seek_help_count.plot(kind="pie",autopct='%2.2f%%')
plt.ylabel(None)
plt.title("Protection for Seeking Resources and Are You Seeking Help")
```
